## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# load as Pandas dataframe
worldbank_json_df = pd.read_json('data/world_bank_projects.json')
worldbank_json_df.head()
```

```python
#print info
worldbank_json_df.info()
```

```python
#1.Find the 10 countries with most projects
worldbank_json_df.countryname.value_counts()[:10]
```

```python
worldbank_json_namecode_df=worldbank_json_df['mjtheme_namecode'].apply(pd.Series)
#[worldbank_json_df['mjtheme_namecode'][i] for i in range(len(worldbank_json_df['mjtheme_namecode']))]
#new_df
worldbank_json_namecode_df
#worldbank_json_namecode_df=pd.melt(worldbank_json_namecode_df)
#worldbank_json_namecode_df=worldbank_json_namecode_df.drop(['variable'],axis=1)
#worldbank_json_namecode_df1=worldbank_json_namecode_df['value'].apply(pd.Series)
#worldbank_json_namecode_df1.groupby('name')['code'].count().sort_values(ascending=False)
```

```python

#melt
worldbank_json_namecode_df1=pd.melt(worldbank_json_namecode_df)
worldbank_json_namecode_df1
```

```python
#Drop the varaible column
worldbank_json_namecode_df1=worldbank_json_namecode_df1.drop(['variable'],axis=1)
worldbank_json_namecode_df1
```

```python
#create new df to extract code and name from dictionary to DF columns as code and name
worldbank_json_namecode_df2=worldbank_json_namecode_df1['value'].apply(pd.Series)
worldbank_json_namecode_df2
```

```python
#worldbank_json_namecode_df2[worldbank_json_namecode_df2['name']==""]=worldbank_json_namecode_df2['code']
worldbank_json_namecode_df2 = worldbank_json_namecode_df2.ffill( axis=1,inplace=False)
worldbank_json_namecode_df2
```

```python
worldbank_json_namecode_df2.dropna()
worldbank_json_namecode_df2
```

```python
#2. Find the top 10 major project themes (using column 'mjtheme_namecode')
#use group by name to get count and apply sort
worldbank_json_namecode_df2.groupby('name')['code'].count().sort_values(ascending=False)
```

```python
#3.In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.
worldbank_json_namecode_df3=worldbank_json_namecode_df2
worldbank_json_namecode_df3.loc[worldbank_json_namecode_df3['name']=='','name']=worldbank_json_namecode_df3.loc[worldbank_json_namecode_df3['name']=='','code']
```

```python
worldbank_json_namecode_df3.groupby('name')['code'].count().sort_values(ascending=False)
```

```python
#another approach
#3.In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.
worldbank_json_namecode_df4=worldbank_json_namecode_df2.copy()
worldbank_json_namecode_df4.apply(lambda x, y: x.where(x != '', y), y=worldbank_json_namecode_df4['code'])
```

```python
worldbank_json_namecode_df4.groupby('name')['code'].count().sort_values(ascending=False)
```
